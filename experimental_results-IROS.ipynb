```python
s = .90625

shape_1a = Shape([Block(0,0,scale=s)])
shape_1b = Shape([Block(0,0,90,scale=s)])
shape_2a = Shape([Block(-.5,0,scale=s),    Block(.5,0,scale=s)])
shape_2b = Shape([Block(0,0.25,scale=s),   Block(0,-0.25,scale=s)])
shape_2c = Shape([Block(-.25,0,90,scale=s),Block(0.25,0,90,scale=s)])
shape_2d = Shape([Block(0,-.5,90,scale=s) ,Block(0,.5,90,scale=s)])
shape_2e = Shape([Block(0,-0.5,90,scale=s),Block(0.25,0.25,0,scale=s)])
shape_2f = Shape([Block(-0.25,0.25,90,scale=s),Block(0.25,-0.25,90,scale=s)])
shape_3a = Shape([Block(-.5,0,scale=s),    Block(.5,0,scale=s),   Block(-1.25,0.25,90,scale=s)])
shape_3b = Shape([Block(-.5,0,scale=s),    Block(.5,0,0,scale=s), Block(-.5,0.5,scale=s)])
shape_3c = Shape([Block(0,0,scale=s),      Block(-1,0,scale=s),   Block(1,0,scale=s)])
shape_3d = Shape([Block(-.5,0,scale=s),    Block(.5,0,scale=s),   Block(-.75,0.75,90,scale=s)])
shape_4a = Shape([Block(-.5,0,scale=s),    Block(.5,0,scale=s),   Block(-1.25,0.25,90,scale=s),Block(1.25,0.25,90,scale=s)])
shape_4b = Shape([Block(-.5,-.25,scale=s), Block(-.5,.25,scale=s),Block(.5,-.25,scale=s),      Block(.5,.25,scale=s)])
shape_6a = Shape([Block(0,0,scale=s),      Block(1.,0,scale=s),   Block(0,0.5,scale=s),        Block(-1.,0.5,scale=s),Block(1.,0.5,scale=s),Block(1.,-.5,scale=s)])
```

```python
tests = {'1_1':         (shape_1b, shape_2c, (s*-0.75, s*0.,     0.),  (s*-0.25,    s*1.0,    0.)),  \
         '1_2':        (shape_2d, shape_1b, (s*-0.5, s*0.5,     0.),  (s*0.0,    s*1.5,    0.)),  \
         '2_1':         (shape_1b, shape_2c, (s*-.75, s*0.0,     0.),  (s*0.,    s*.75,   90.)), \
         '2_2':        (shape_2e, shape_1b, (s*-0.5, s*0.5,     0.),  (s*0,    s*1.5,  0.)),  \
         '3_1':        (shape_1b, shape_2c, (s*-0.75, s*0.,     0.),  (s*-0.75,    s*0.5,    0.)),  \
         '3_2':        (shape_2f, shape_1b, (s*-0.75, s*0.25,     0.),  (s*0.25,    s*0.75,    0))
        }
```

```python
name_template = 'results/IROS/2-21/shape_shifting_%s_0%d.csv'

o_x = 2.85
o_y = 1.05
offsets = {
            '1_1':[o_x,        o_y, np.pi/2],
            '1_2':[o_x+0.225,  o_y, np.pi/2],
            '2_1':[o_x,        o_y, np.pi/2],
            '2_2':[o_x+0.5625, o_y-0.1125, np.pi],
            '3_1':[o_x+0.02,   o_y, np.pi/2],
            '3_2':[o_x+0.10,   o_y, np.pi/2]
}

exps = {test:{'data': []} for test in tests}

for test, x in exps.items():
    print test
    calc = shapeshift_trajectory(*tests[test], N=11, boat_type=ShapeBoat_spline, boat_type_init=ShapeBoat_spline)
    x['S'] = calc['S']
    x['boat'] = calc['boat']
    for i in range(3):
        try:
            exp = np.genfromtxt(name_template % (test,i+1),delimiter=',')[:,1:]
            exp[:,:3]  -= offsets[test]
            exp[:,6:9] -= offsets[test]
            inds1 = exp[:,2]-exp[:,8]>np.pi
            inds2 = exp[:,2]-exp[:,8]<-np.pi
            exp[:,2][inds1] -= 2*np.pi
            exp[:,2][inds2] += 2*np.pi   
            x['data'].append({'vals': exp[:, :6], 'refs':exp[:, 6:12]})
        except:
            pass
```

```python
for test, x in exps.items():
    angle_avgs = []
    pos_avgs = []
    print test
    for trial in x['data']:
        error = (trial['vals']-trial['refs'])**2
        pos_avg = np.average((error[:,0]+error[:,1])**0.5)
        angle_avg = np.average(error[:,2]**0.5*180/np.pi)
        angle_avgs.append(angle_avg)
        pos_avgs.append(pos_avg)
        print 'avg_position_error', pos_avg
        print 'avg_angle_error', angle_avg
    print bold('all avg_position_error: %.3f' % (sum(pos_avgs)/len(pos_avgs)))
    print bold('all avg_angle_error: %.1f' % (sum(angle_avgs)/len(angle_avgs)))
    print 
```

```python
def tracking_plot(trial):
    vals = trial['vals']
    refs = trial['refs']
    plt.figure(figsize=(6,4))
    plt.plot(np.arange(len(vals))/5., vals[:,0], 'C0--')
    plt.plot(np.arange(len(vals))/5., refs[:,0], 'C0', label='x(m)')
    plt.plot(np.arange(len(vals))/5., vals[:,1], 'C1--')
    plt.plot(np.arange(len(vals))/5., refs[:,1], 'C1', label='y(m)')
    plt.plot(np.arange(len(vals))/5., np.max(np.vstack((-vals[:,2],vals[:,2])), axis=0), 'C2--')
    plt.plot(np.arange(len(vals))/5., refs[:,2], 'C2', label='theta(rad)')
    plt.legend()
    plt.xlabel('time(s)')
    plt.show()
    
tracking_plot(exps['3_2']['data'][0])
```

```python
style.use('default')

style.use('tableau-colorblind10')

print plt.rcParams['axes.prop_cycle'].by_key()['color']

from matplotlib.collections import PathCollection

fig_name_template = 'results/IROS/shape_shifting_%s.pdf'

colors = [x for x in plt.rcParams['axes.prop_cycle'].by_key()['color'] if x not in ['#ff7f0e','#FF800E','#C85200', '#FFBC79']]

for test,x in exps.items(): 
    print test
    fig, ax = plt.subplots(nrows=1,ncols=1, figsize=(10,10))    
    ax.set_prop_cycle(color=colors)
    x['boat'].plot_hulls(plot=(fig,ax), text=False)
    plt.plot(x['data'][0]['refs'][1:,0],x['data'][0]['refs'][1:,1], color='black', linewidth='8')
    for trial in x['data']:
        plt.plot(trial['vals'][10:,0],trial['vals'][10:,1], color='darkorange', linewidth='5',  alpha=0.6)
    
    r = transforms.Affine2D().rotate_deg(90)
    for y in ax.images + ax.lines + ax.collections:
        trans = y.get_transform()
        y.set_transform(r+trans)
        
    for p in ax.patches:
        h,w = p.get_height(), p.get_width()
        p.set_height(w)
        p.set_width(h)
        x,y = p.get_xy()
        p.set_xy((-y,x-w/2))
  
    xlim, ylim = ax.get_xlim(), ax.get_ylim()
    ax.set_ylim((xlim[0],xlim[1]))
    ax.set_xlim((-ylim[1],-ylim[0]))
    
    plt.show()
    fig.savefig(fig_name_template % test, bbox_inches='tight')
```
